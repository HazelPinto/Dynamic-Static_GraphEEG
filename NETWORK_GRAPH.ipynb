```python
import pandas as pd
import matplotlib.pyplot as plt
import networkx as nx
```

```python
df = pd.read_csv(r'Subject17_1.csv')
column_names = list(df.columns.values)
column_names
```

```python
# Static graph 
datos = np.loadtxt('Subject17_1.csv', delimiter=',')


similitud = np.corrcoef(datos.T)


red = nx.Graph()

for i in column_names:
    red.add_node(i)

for i in range(len(datos[0])):
    for j in range(i+1, len(datos[0])):
        if similitud[i,j] > 0.8:
            red.add_edge(column_names[i],column_names[j])
            
pos = nx.spring_layout(red, scale=20, k=3/np.sqrt(red.order()))
from pylab import rcParams
d = dict(red.degree)

fig = plt.figure(figsize=(10, 10))
plt.title("Subject17_1")

nx.draw(red, pos, node_color='yellow', 
        with_labels=True, 
        nodelist=column_names, 
        node_size=[d[k]*100 for k in d])

plt.show()
```

# Dynamic Graph

```python
import numpy as np
import networkx as nx
from pyvis.network import Network

# Load data from CSV file
datos = np.loadtxt('Subject17_1.csv', delimiter=',')
column_names = [f"Node{i}" for i in range(len(datos))]

# Calculate similarity matrix
similitud = np.corrcoef(datos.T)

# Create a new Pyvis Network object
red_dinamica = Network(width="100%", height="800px", directed=False, notebook=True)

# Add nodes to the network
for i in column_names:
    red_dinamica.add_node(i)

# Add edges to the network based on similarity matrix
for i in range(len(datos[0])):
    for j in range(i+1, len(datos[0])):
        if similitud[i,j] > 0.8:
            red_dinamica.add_edge(column_names[i],column_names[j], title=f"similarity={similitud[i,j]}")

# Set the layout of the network
red_dinamica.force_atlas_2based()

# Display the network
red_dinamica.show("Subject17_1.html")
```

```python


# Load the data
data = np.loadtxt('Subject17_1.csv', delimiter=',')

# Define the columns
df = pd.read_csv(r'Subject17_1.csv')
column_names = list(df.columns.values)

# Create an empty network
graph = nx.Graph()

# Add nodes to the network
for i in column_names:
    graph.add_node(i)

# Define the initial threshold
threshold = 0.8

# Calculate the initial correlation
similarity = np.corrcoef(data.T)

# Add connections that meet the initial threshold
for i in range(len(column_names)):
    for j in range(i + 1, len(column_names)):
        if similarity[i, j] > threshold:
            graph.add_edge(column_names[i], column_names[j])

# Draw the initial network
pos = nx.spring_layout(graph, scale=20, k=3/np.sqrt(graph.order()))
degree = dict(graph.degree)
nx.draw(graph, pos, node_color='yellow',
        with_labels=True,
        nodelist=column_names,
        node_size=[degree[k]*100 for k in degree])
plt.title("Subject17_1")
plt.show()

# Update the network dynamically
for t in range(1, len(data)):
    # Calculate the correlation at time t
    similarity_t = np.corrcoef(data[:t+1, :].T)

    # Get connections that meet the threshold
    connections = [(column_names[i], column_names[j]) for i in range(len(column_names)) 
                   for j in range(i+1, len(column_names)) if similarity_t[i, j] > threshold]

    # Add new connections
    graph.add_edges_from(connections)

    # Remove connections that no longer meet the threshold
    to_remove = [e for e in graph.edges() if e not in connections]
    graph.remove_edges_from(to_remove)

    # Draw the updated network
    pos = nx.circular_layout(graph)
    degree = dict(graph.degree)
    nx.draw(graph, pos, node_color='yellow', with_labels=True, nodelist=column_names, 
            node_size=[degree[k]*100 for k in degree], font_size=8.7, font_color='black', edge_color='black', width=1)
    plt.title("Subject17_1 - Time Step {}".format(t))
    plt.savefig("Subject17_1 - Time Step {}.png".format(t))
    plt.show()
```

```python

datos = np.loadtxt('Subject17_2.csv', delimiter=',')

# Definir las columnas

column_names = list(df.columns.values)

# Crear una red vacÃ­a
red = nx.Graph()

# Agregar los nodos a la red
for i in column_names:
    red.add_node(i)

# Definir el umbral inicial
threshold = 0.8

# Calcular la correlaciÃ³n inicial
similitud = np.corrcoef(datos.T)

# Agregar las conexiones que cumplen el umbral inicial
for i in range(len(column_names)):
    for j in range(i+1, len(column_names)):
        if similitud[i,j] > threshold:
            red.add_edge(column_names[i],column_names[j])

# Dibujar la red inicial
pos = nx.spring_layout(red, scale=20, k=3/np.sqrt(red.order()))
d = dict(red.degree)
nx.draw(red, pos, node_color='lightblue', 
        with_labels=True, 
        nodelist=column_names, 
        node_size=[d[k]*100 for k in d])
plt.title("Subject17_1")
plt.show()

# Actualizar la red dinÃ¡micamente
for t in range(1, len(datos)):
    # Calcular la correlaciÃ³n en el tiempo t
    similitud_t = np.corrcoef(datos[:t+1,:].T)

    # Obtener las conexiones que cumplen el umbral
    connections = [(column_names[i],column_names[j]) for i in range(len(column_names)) 
                   for j in range(i+1, len(column_names)) if similitud_t[i,j] > threshold]

    # Agregar las nuevas conexiones
    red.add_edges_from(connections)

    # Eliminar las conexiones que ya no cumplen el umbral
    to_remove = [e for e in red.edges() if e not in connections]
    red.remove_edges_from(to_remove)

    # Dibujar la red actualizada
    pos = nx.circular_layout(red)
    d = dict(red.degree)
    nx.draw(red, pos, node_color='lightblue', with_labels=True, nodelist=column_names, node_size=[d[k]*100 for k in d], font_size=8.7, font_color='black', edge_color='black', width=1)
    plt.title("Subject17_2 - Time Step {}".format(t))
    plt.savefig("Subject17_2 - Time Step {}.png".format(t))
    plt.show()

```
